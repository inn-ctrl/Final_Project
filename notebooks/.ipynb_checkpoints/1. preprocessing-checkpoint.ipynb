Dimensionality Reduction

```python
Applying PCA to the encoded dataset gives us the features that mostly contribute to employee attrition
```

```python
#separating features from target variable 'Attrition'
X = df.drop('Attrition', axis = 1)
y = df['Attrition']

#standardizing features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

#evr and sorting
explained_variance_ratio = pca.explained_variance_ratio_
explained_variance_ratio_sorted = sorted(explained_variance_ratio, reverse = True)

#principal components
max_var_component_index = explained_variance_ratio.argmax()

#feature wieghts
feature_loadings = pca.components_[max_var_component_index]

#feature loadings in a dataframe
components_df = pd.DataFrame({'Feature': X.columns, 'Loading': feature_loadings})
components_df = components_df.reindex(components_df['Loading'].abs().sort_values(ascending = False).index)
```

```python
#we plot the above on a graph and only extract the components that highly affect the attrition
plt.figure(figsize = (12, 5))
plt.bar(components_df['Feature'], components_df['Loading'])
plt.xticks(rotation=90)
plt.show()
```

Based on the findings above, we make an arbitrary decision to only keep 17 features that has a visibly high explained variance ration compared to attrition
